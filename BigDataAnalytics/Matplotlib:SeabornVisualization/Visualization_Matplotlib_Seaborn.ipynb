```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('elections (1).csv')

# Convert 'Year' column to integers
df['Year'] = df['Year'].astype(int)
plt.figure(figsize=(15, 8))

# Create the line plot
sns.lineplot(
    x='Year', y='%', hue='Party', data=df,
    palette='tab20', linewidth=2.5, marker='o'
)

# Customize the plot
plt.xlabel('Election Year', fontsize=12)
plt.ylabel('Vote Percentage (%)', fontsize=12)
plt.title('Party Performance Over Time', fontsize=14)
plt.legend(title='Political Party', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()



```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('elections (1).csv')

# Filter only winning candidates
df_winners = df[df['Result'] == 'win']

# Count the number of wins per party
df_grouped = df_winners['Party'].value_counts().reset_index()
df_grouped.columns = ['Party', 'Wins']
df_grouped = df_grouped.sort_values(by='Wins', ascending=False)

plt.figure(figsize=(14, 7))
ax = sns.barplot(x='Party', y='Wins', data=df_grouped, palette='tab20')

# Adjust the y-axis limit to give space for the label
plt.ylim(0, df_grouped['Wins'].max() + 1)

# Add data labels
for index, row in df_grouped.iterrows():
    ax.text(index, row['Wins'] + 0.2, row['Wins'], ha='center', fontsize=10)

plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Political Party', fontsize=12)
plt.ylabel('Number of Wins', fontsize=12)
plt.title('Number of Wins per Political Party', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('elections (1).csv')
df_winners = df[df["Result"] == "win"]

# Convert 'Year' to integer for sorting
df_winners['Year'] = df_winners['Year'].astype(int)
df_winners = df_winners.sort_values("Year")

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="%", data=df_winners, marker="o", color="blue", linewidth=2.5)

plt.xticks(rotation=45)
plt.xlabel("Election Year")
plt.ylabel("Winning Popular Vote Percentage")
plt.title("Trend of Popular Vote Percentage for Winning U.S. Presidential Candidates (1800 - 2000)")
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('elections (1).csv')

# select the top 10 candidates by popular vote
df_sorted = df.dropna(subset=["Popular vote"]).sort_values(by="Popular vote", ascending=False).head(10)

plt.figure(figsize=(12,6))
# bar plot with a custom palette
ax = sns.barplot(x="Popular vote", y="Candidate", data=df_sorted,
                 palette=sns.color_palette("colorblind", len(df_sorted)))

# format the x-axis tick labels to add comma separators
ax.set_xticklabels([f"{int(x):,}" for x in ax.get_xticks()])

plt.title("Top 10 Most Popular Candidates by Vote Count")
plt.xlabel("Popular Vote")
plt.ylabel("Candidate")

for p in ax.patches:
    width = p.get_width()
    ax.text(width + width*0.01, p.get_y() + p.get_height()/2, f'{int(width):,}',
            ha='left', va='center')

plt.show()

```

```python
import zipfile
import os

zip_path = "/content/namesbystate.zip"
extract_path = "/content/namesbystate"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

print(os.listdir(extract_path))
```

```python
import pandas as pd

columns = ["State", "Sex", "Year", "Name", "Count"]
babynames = pd.DataFrame(columns=columns)

# loop through all TXT files in the extracted folder
for file in os.listdir(extract_path):
    if file.endswith(".TXT"):
        file_path = os.path.join(extract_path, file)
        state_df = pd.read_csv(file_path, header=None, names=columns)
        babynames = pd.concat([babynames, state_df], ignore_index=True)
babynames.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# filter data for Mary in TX, CA, and FL
states_of_interest = ["TX", "CA", "FL"]
mary_trend = babynames[(babynames["Name"] == "Mary") & (babynames["State"].isin(states_of_interest))]

plt.figure(figsize=(12, 6))
sns.lineplot(data=mary_trend, x="Year", y="Count", hue="State",
             palette="colorblind", linewidth=2.5, ci=None)

plt.title("Popularity of the Name 'Mary' Over Time in TX, CA, and FL", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="State", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()





```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# filter for California only
ca_names = babynames[babynames["State"] == "CA"].copy()
ca_names["Count"] = pd.to_numeric(ca_names["Count"], errors="coerce")

# find the 10 most common names
top_ca_names = ca_names.groupby("Name")["Count"].sum().nlargest(10).reset_index()
sns.set_context("talk")

# plot
plt.figure(figsize=(12, 5))
ax = sns.barplot(data=top_ca_names, x="Count", y="Name", palette="colorblind")
plt.title("Top 10 Most Common Baby Names in California", fontsize=14)
plt.xlabel("Total Count", fontsize=12)
plt.ylabel("Name", fontsize=12)

# adjust axis tick label sizes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis="x", linestyle="--", alpha=0.7)

# change the size of the numbers at the end of the bars
for index, row in top_ca_names.iterrows():
    ax.text(row["Count"] + 50, index, str(row["Count"]), color="black", va="center", fontsize=8)  # Smaller font size

# Show the plot
plt.show()



```

```python
plt.figure(figsize=(12, 6))
gender_trend = babynames[babynames["State"] == "WY"].groupby(["Year", "Sex"])["Count"].sum().reset_index()
sns.lineplot(data=gender_trend, x="Year", y="Count", hue="Sex", linewidth=3, palette=["blue", "pink"])

plt.title("Gender Distribution of Baby Names Over Time in Wyoming", fontsize=16)
plt.xlabel("Year", fontsize=13)
plt.ylabel("Total Count", fontsize=13)
plt.legend(title="Sex", loc="upper left")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
gender_trend = babynames[babynames["State"] == "WY"].groupby(["Year", "Sex"])["Count"].sum().reset_index()
gender_trend_pivot = gender_trend.pivot(index="Year", columns="Sex", values="Count")
gender_trend_pivot.plot.area(stacked=True, color=["blue", "pink"], alpha=0.6, figsize=(12, 6))

plt.title("Gender Proportions of Baby Names in Wyoming Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Total Count", fontsize=14)
plt.legend(title="Sex", loc="upper left")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
